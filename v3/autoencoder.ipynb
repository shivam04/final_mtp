```python
import pandas as pd
import numpy as np
import glob, os, sys
import time
import pickle
import datetime
```

```python
df = pd.read_csv('finalset/patient_data_48_hours.csv')
df.head(50)
```

```python
matrix3D = np.array(df.drop(['SUBJECT_ID', 'TimeStamp'], 1))
matrix3D
```

```python
matrix3D.shape
```

```python
matrix3D = np.array(matrix3D).reshape((6587, 49, 37))
matrix3D.shape
```

```python
outcomes = pd.read_csv('finalset/outcomes.csv')
outcomes.head(10)
```

```python
Y = np.array(outcomes.drop(['SUBJECT_ID'], 1))
Y.shape
```

```python
X = matrix3D
X.shape
```

```python
X_train = X[:5270]
X_test = X[5270:]
Y_train = Y[:5270]
Y_test = Y[5270:]
display(X_train.shape)
display(X_test.shape)
display(Y_train.shape)
display(Y_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout,RepeatVector,TimeDistributed
from keras.utils import plot_model
```

```python
model = Sequential()
model.add(LSTM(100, activation='relu', input_shape=(X_train.shape[1],X_train.shape[2])))
model.add(RepeatVector(X_train.shape[1]))
model.add(LSTM(100, activation='relu', return_sequence=True))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
print(model.summary())
model.fit(X_train, Y_train, validation_data=(X_test, Y_test), epochs=10,batch_size=50)
```

```python
scores = model.evaluate(X_test, Y_test)
scores[1]*100
```

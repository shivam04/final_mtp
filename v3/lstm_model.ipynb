```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
ans = []
class LSTM(object):
    def __init__(self):
        # define lower bound of benchmark function
        self.Lower = 0
        # define upper bound of benchmark function
        self.Upper = 1

    # function which returns evaluate function
    def function(self):
        def evalute(D,sol):
            from keras.models import Sequential
            from keras.layers import Dense
            from keras.layers import LSTM
            from keras.layers import Dropout
            sol = np.array(sol)
            sol = sigmoid(sol)
            #display(sol)
            op = sol>=0.5
            X_tr = X_train[:,:,op]
            X_te = X_test[:,:,op]
            model = Sequential()
            model.add(LSTM(100, input_shape=(X_tr.shape[1],X_tr.shape[2])))
            model.add(Dropout(0.2))
            model.add(Dense(100,activation='tanh'))
            model.add(Dense(25,activation='tanh'))
            model.add(Dense(10,activation='tanh'))
            model.add(Dense(1, activation='sigmoid'))
            model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
            model.fit(X_tr, Y_train, validation_data=(X_te, Y_test), epochs=30,batch_size=50,verbose=0)
            scores = model.evaluate(X_te, Y_test, verbose=0)
            k = -1*scores[1]
            return k
        return evalute
```

```python
from NiaPy.algorithms.basic import BatAlgorithm
for i in range(10):
    algorithm = BatAlgorithm(D=37, NP=370, nFES=10, benchmark=LSTM())
    best = algorithm.run()
    print(-1*best[1])
    ans.append(-1*best[1])
```

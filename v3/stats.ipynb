```python
import math
from datetime import datetime
import time
g = 0
ww = []
for k in np.array(defd['DOD']):
    #print(k)
    try:
        if math.isnan(k):
            g+=1
            ww.append(0)
    except:
        d = datetime.strptime(k, "%Y-%m-%d %H:%M:%S")
        ww.append(time.mktime(d.timetuple()))
        
g
```

```python
pp = []
for k in np.array(defd['DOB']):
    #print(k)
    d = datetime.strptime(k, "%Y-%m-%d %H:%M:%S")
    try:
        pp.append(time.mktime(d.timetuple()))
    except:
        pp.append(0)
pp
```

```python
len(pp)
```

```python
ages = 0
for i in range(0,len(pp)):
    if pp[i]==0:
        ages+=0
    elif ww[i] == 0:
        y = (pp[i]/(60*60*24*365))
        ages+=y
    else:
        y = ((ww[i]-pp[i])/(60*60*24*365))
        ages+=y
ages
```

```python
avg_age = ages/(6587.0)
```
